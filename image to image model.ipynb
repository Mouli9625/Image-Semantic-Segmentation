```python
import torch
from diffusers import StableDiffusionXLImg2ImgPipeline
from diffusers.utils import load_image
import os  # For path manipulation

pipe = StableDiffusionXLImg2ImgPipeline.from_pretrained(
    "stabilityai/stable-diffusion-xl-refiner-1.0", torch_dtype=torch.float16, variant="fp16", use_safetensors=True
)
pipe = pipe.to("cuda")
def generate_from_file(prompt_file, init_image_path, output_dir):
    """
    This function reads prompts from a text file, generates images for each prompt,
    and saves them to the specified output directory.

    Args:
        prompt_file (str): Path to the text file containing prompts.
        init_image_path (str): Path to the initial image for image-to-image diffusion.
        output_dir (str): Path to the directory where generated images will be saved.
    """

    init_image = load_image(init_image_path).convert("RGB")
    with open(prompt_file, 'r') as f:
        for line in f:
            prompt = line.strip()  # Remove leading/trailing whitespace

            # Generate the image
            image = pipe(prompt, image=init_image).images[0]

            # Create a unique filename based on the prompt
            filename = f"{prompt.replace(' ', '_')}.jpg"  # Replace spaces with underscores

            # Create the output directory if it doesn't exist
            os.makedirs(output_dir, exist_ok=True)  # Safe directory creation

            # Save the generated image
            image.save(os.path.join(output_dir, filename))
            print(f"Generated image for prompt: {prompt} and saved as {filename}")


# Replace these with your actual file paths
prompt_file = "/content/trail.txt"
init_image_path = "/content/rcc-road-dividers.jpg"
output_dir = "/content/generated_images1"  # Make sure this directory exists or has write permissions

generate_from_file(prompt_file, init_image_path, output_dir)
```
